## Import libraries

```python
import argparse
import torch.nn as nn
import torch
from torch.autograd import Variable
import torchvision.transforms as transforms
from torchvision.utils import save_image
import numpy as np
import os
import pickle
```

## Configure training parameters

```python
# Function to load the saved opt variable
def load_opt():
    with open('opt.pkl', 'rb') as f:
        return pickle.load(f)

# Load and print the opt variable
opt = load_opt()
```

```python
import torch
import numpy as np
from torchvision.utils import save_image
import os

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()

        self.init_size = opt.img_size // 4
        self.l1 = nn.Sequential(nn.Linear(opt.latent_dim, 128 * self.init_size ** 2))

        self.conv_blocks = nn.Sequential(
            nn.BatchNorm2d(128),
            nn.Upsample(scale_factor=2),
            nn.Conv2d(128, 128, 3, stride=1, padding=1),
            nn.BatchNorm2d(128, 0.8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Upsample(scale_factor=2),
            nn.Conv2d(128, 64, 3, stride=1, padding=1),
            nn.BatchNorm2d(64, 0.8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, opt.channels, 3, stride=1, padding=1),
            nn.Tanh(),
        )

    def forward(self, z):
        out = self.l1(z)
        out = out.view(out.shape[0], 128, self.init_size, self.init_size)
        img = self.conv_blocks(out)
        return img
    
# Instancia el generador
generator = Generator()

# Cargar los parmetros del modelo guardado
checkpoint = torch.load('V1/models/model_epoch_1000.pth', weights_only=True)
generator.load_state_dict(checkpoint['generator_state_dict'], strict=False)

# Generar nuevas imgenes
def generate_images(generator, num_images, latent_dim, clase, output_dir='generated_images'):
    generator.eval()  # Pone el modelo en modo evaluacin
    os.makedirs(output_dir, exist_ok=True)
    
    for i in range(num_images):
        z = torch.FloatTensor(np.random.normal(0, 1, (1, latent_dim)))
        gen_img = generator(z)
        save_image(gen_img.data, f"{output_dir}/{clase}_{i + 1}.png", normalize=True)
        print(f"Imagen {i + 1} guardada en {output_dir}/{clase}_{i + 1}.png")

# Define el nmero de imgenes a generar y la dimensin latente
num_images_to_generate = 5479
latent_dim = opt.latent_dim
clase="PITUITARY"

# Genera y guarda las imgenes
generate_images(generator, num_images_to_generate, latent_dim, clase)
```
